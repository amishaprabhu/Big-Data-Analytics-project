```python
#according to gender

M_f=pd.DataFrame(df["Gender"].value_counts().reset_index(drop=False))

fig=px.pie(M_f,values="Gender",names=M_f["index"],title="Who among male And Female is most interested in travelling",
          color_discrete_sequence=["#4A4F49","#959E92"],width=600,height=450)

fig.show()
```

```python
#according to ages

age=pd.DataFrame(df["Age"].value_counts().reset_index(drop=False).rename(columns={"index":"Age","Age":"Count"}))

fig=px.pie(age,values=age['Count'],names="Age",title="What age people are more interested in travelling",
                          color_discrete_sequence=px.colors.sequential.RdBu,width=600,height=450)

fig.show()
```

```python
#Favorite Tourism Destination in India

india=pd.DataFrame(df["FavoriteTourismDest_india"].value_counts().reset_index().rename(columns={"index":"FavoriteTourismDest_india","FavoriteTourismDest_india":"count"}))

fig=px.bar(india,x="count",y="FavoriteTourismDest_india",title="Favorite Tourism Destination in India",color_discrete_sequence=['lightblue'], text='count')

fig.update_xaxes(showgrid=False)

fig.update_yaxes(showgrid=False, categoryorder='total ascending', ticksuffix=' ', showline=False)

fig.update_traces(hovertemplate=None, marker=dict(line=dict(width=0)))

fig.update_layout(margin=dict(t=80, b=0, l=70, r=40),hovermode="y unified",
                  xaxis_title=' ', yaxis_title=" ", height=400,plot_bgcolor='#333', paper_bgcolor='#333',
title_font=dict(size=25, color='#8a8d93', family="Lato, sans-serif"),
font=dict(color='#8a8d93'),
                  legend=dict(orientation="h", yanchor="bottom", y=1, xanchor="center", x=0.5),
                  hoverlabel=dict(bgcolor="black", font_size=13, font_family="Lato, sans-serif"))
```

```python
#Favorite Tourism Destination in abroad

faran=pd.DataFrame(df["FavoriteTourismDest_abroad"].value_counts().reset_index().rename(columns={"index":"FavoriteTourismDest_abroad","FavoriteTourismDest_abroad":"count"}))

fig=px.bar(faran,x="count",y="FavoriteTourismDest_abroad",title="Favorite Tourism Destination in abroad",color_discrete_sequence=['lightblue'], text='count')

fig.update_xaxes(showgrid=False)

fig.update_yaxes(showgrid=False, categoryorder='total ascending', ticksuffix=' ', showline=False)

fig.update_traces(hovertemplate=None, marker=dict(line=dict(width=0)))

fig.update_layout(margin=dict(t=80, b=0, l=70, r=40),hovermode="y unified",
                  xaxis_title=' ', yaxis_title=" ", height=400,plot_bgcolor='#333', paper_bgcolor='#333',
title_font=dict(size=25, color='#8a8d93', family="Lato, sans-serif"),
font=dict(color='#8a8d93'),
                  legend=dict(orientation="h", yanchor="bottom", y=1, xanchor="center", x=0.5),
                  hoverlabel=dict(bgcolor="black", font_size=13, font_family="Lato, sans-serif"))
```

```python
#According to City peoples like to visit to which tourist Destination In India

fig = px.bar(df,
               y    = 'FavoriteTourismDest_india',
               x    = "City",
               orientation = 'v',
               color       = 'City',
               hover_name  = 'City',
               title       = 'Which city peoples like to visit to which tourist Des..',
               opacity     = 0.8,
               color_continuous_scale = px.colors.sequential.Viridis,
               height = 450
            )

fig.update_layout(showlegend=False)
fig.update_xaxes(title='Cities')
fig.update_layout(margin=dict(t=80, b=0, l=70, r=40),hovermode="y unified",
                  xaxis_title=' ', yaxis_title=" ", height=450,plot_bgcolor='#333', paper_bgcolor='#808080')

fig.update_yaxes(title='FavoriteTourismDest_india',showticklabels=False)
fig.show()
```

```python
#According to City peoples like to visit to which tourist Destination In abroad


fig=px.bar(df,
          x="FavoriteTourismDest_abroad",
          y="City",
          orientation="v",
          color="City",
          hover_name="City",
          title="Which City peoples like to visit in abroad",
          opacity=0.8,
           color_continuous_scale =px.colors.qualitative.G10,
               height = 450)
fig.update_layout(showlegend=False)
fig.update_xaxes(title="Cities")
fig.update_layout(margin=dict(t=80, b=0, l=70, r=40),hovermode="y unified",
                  xaxis_title=' ', yaxis_title=" ", height=400,plot_bgcolor='#333', paper_bgcolor='#808080')

fig.update_yaxes(title="FavoriteTourismDest_abroad",showticklabels=False)
fig.show()
```

```python
#Travelling according to Education in india & abroad

trace=go.Bar(
x=df["Education"],
y=df["FavoriteTourismDest_india"],
marker=dict(color="rgb(120,50,800)",
           line=dict(color="rgb(200,210,220)",width=1.2)))



layout=go.Layout(template="plotly_dark",title="Travelling according to Education in India:",xaxis=dict(title="Education",tickangle=45),yaxis=dict(title="Cities"))
fig=go.Figure(data=[trace],layout=layout)
fig.show()

trace_second=go.Bar(
x=df["Education"],
y=df["FavoriteTourismDest_abroad"],
marker=dict(color="rgb(120,50,800)",
           line=dict(color="rgb(200,210,220)",width=1.2)))



layout=go.Layout(template="plotly_dark",title="Travelling according to Education in abroad:",xaxis=dict(title="Education",tickangle=45),yaxis=dict(title="Cities"))
fig=go.Figure(data=[trace_second],layout=layout)
fig.show()
```
